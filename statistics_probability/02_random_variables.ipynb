### 1. Discrete Random Variable <a name='DiscreteRV'></a>
**Definition:**  
A **discrete random variable (RV)** takes **countable values**, often integers, with associated probabilities.  

**Examples:**  
- Number of heads in 3 coin tosses → {0, 1, 2, 3}  
- Number of cars passing a street in 1 hour → {0, 1, 2, ...}  

---

### 2.Continuous Random Variable <a name='ContinuousRV'></a>

**Definition:**  
A **continuous random variable (RV)** takes values from a **continuous range** (any real number within an interval) and is described by a **Probability Density Function (PDF)**.  

**Examples:**  
- Height of students → any value in a range  
- Temperature → any value in a continuous interval  

---


### Probability Distribution Function (PDF / PMF) <a name='PDF'></a>

**Definition:**  
- **PMF (Probability Mass Function):** For **discrete** random variables, gives the probability of each possible value.  
- **PDF (Probability Density Function):** For **continuous** random variables, describes the likelihood of taking values in a small interval.

**Key Points:**  
- PMF: $P(X = x_i), \sum_i P(X = x_i) = 1$  
- PDF: $P(a \le X \le b) = \int_a^b f_X(x) dx$, $\int_{-\infty}^{\infty} f_X(x) dx = 1$

---

```python
# Python Example: PMF (Discrete)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Discrete RV: number of heads in 3 coin tosses
n, p = 3, 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf)
plt.title("PMF: Discrete Random Variable")
plt.xlabel("Number of heads")
plt.ylabel("Probability")
plt.show()
```

```python
### Python Example: PDF (Continuous)
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Continuous RV: standard normal
x = np.linspace(-4, 4, 200)
pdf = norm.pdf(x, loc=0, scale=1)

plt.plot(x, pdf)
plt.title("PDF: Continuous Random Variable")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

## Cumulative Distribution Function (CDF) <a name='CDF'></a>
**Definition:**  
The **Cumulative Distribution Function (CDF)** gives the probability that a random variable **X** is less than or equal to a certain value **x**:

$$
F_X(x) = P(X \le x)
$$

- For **discrete RVs:** sum of PMF up to x  
- For **continuous RVs:** integral of PDF up to x  

---

```python
# Python Example: Discrete CDF
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Discrete RV: number of heads in 3 coin tosses
n, p = 3, 0.5
x = np.arange(0, n+1)
cdf = binom.cdf(x, n, p)

plt.step(x, cdf, where='post', color='blue')
plt.title("CDF: Discrete Random Variable")
plt.xlabel("Number of heads")
plt.ylabel("Cumulative Probability")
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Python Example: Continuous CDF
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Continuous RV: standard normal
x = np.linspace(-4, 4, 200)
cdf = norm.cdf(x, loc=0, scale=1)

plt.plot(x, cdf, color='red')
plt.title("CDF: Continuous Random Variable")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.show()
```

## 5.Mean / Expected Value <a name='Mean'></a>
**Definition:**  
The **expected value (mean)** of a random variable is the **long-term average value** it would take if the experiment were repeated many times.

- **Discrete RV:**  
$$
E[X] = \sum_i x_i \cdot P(X=x_i)
$$

- **Continuous RV:**  
$$
E[X] = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx
$$

**Interpretation:**  
- Gives the "center" or "balance point" of the probability distribution.  
- Key for **data summarization, predictions, and risk analysis**.

---


```python
#Continuous RV
from scipy.stats import norm

# Continuous RV: standard normal
mean_continuous = norm.mean(loc=2, scale=1)
print("Expected value (mean) of continuous RV:", mean_continuous)
```

## 6.Histogram for Data <a name='Histogram'></a>
**Definition:**  
A **histogram** is a graphical representation of the **distribution of a dataset**, approximating the probability distribution of a random variable.  
- Divides data into **bins** (intervals)  
- Height of each bin shows the **frequency or density** of data in that interval  

**Uses in Data Science:**  
- Visualize distribution of data  
- Identify skewness, spread, and outliers  
- Approximate PDF from real data  

---

```python
# Python Example: Histogram
import numpy as np
import matplotlib.pyplot as plt

#Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)
plt.figure(figsize=(7,4))
plt.hist(data, bins=20, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
