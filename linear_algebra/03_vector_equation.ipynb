## 1. Vectors<a name='Vectors'>
A **vector** is an **ordered list of numbers** representing a point or direction in space:

$$
\mathbf{v} = 
\begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix} \in \mathbb{R}^n
$$
Vectors can be **added**, **scaled**, and used in **linear combinations.**
___
**Example:**

```python
import numpy as np
v = np.array([2, 3, 4])
print("Vector v:", v)
```

## 2. Vector Equations<a name='VectorEquations'></a>
A vector equation expresses a system of linear equations as a linear combination of vectors:
$$ x_1v_1 + x_2v_2 + ... + x_nv_n = b$$
- $x_1,x_2,...,x_n$ are **scalars**
- $v_1,v_2,...,v_n$ are **vectors**
___
**Example:**

```python
v1 = np.array([1, 0])
v2 = np.array([0, 1])
x1, x2 = 2, 3

b = x1*v1 + x2*v2
print("Vector equation result:", b)
```

## 3. Span of Vectors<a name='SpanofVectors'></a>
The span of vectors is the set of all possible linear combinations:
$$\mathbf{Span} \{v_1, v_2\} = \{x_1v_1 + x_2v_2 | x_1, x_2  \in R\}$$

```python
# v1 = [1,0], v2 = [0,1] span the 2D plane
# Any vector b = [x, y] can be written as a combination of v1 and v2
b = np.array([5, -2])
# Solve for coefficients
coeffs = np.linalg.solve(np.column_stack((v1, v2)), b)
print("Coefficients to express b:", coeffs)
```

## 4. Linear Independence<a name= 'LinearIndependence'></a>
A set of vectors $\{ v_1 ,... ,v_n\}$ is linearly independent if:
$$ c_1v_1 + c_2v_2 + ... + c_nv_n = 0 \quad -> c_1=c_2=...=c_n=0$$
Otherwise, the vectors are linearly dependent.
<br>
Linear independence is essential to determine the basis of a vector space and the dimension.

```python
import numpy as np
V = np.array([[1, 0, 1],
              [0, 1, 1],
              [0, 0, 0]]).T  # 3 vectors

rank = np.linalg.matrix_rank(V)
print("Rank of matrix:", rank)
print("Number of vectors:", V.shape[1])

if rank == V.shape[1]:
    print("Vectors are linearly independent")
else:
    print("Vectors are linearly dependent")
```

## 5. Visualization — Span and Linear Independence<a name= 'Visualization'></a>

```python
# Visualization — Span and Linear Independence
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
v1 = np.array([2, 1])
v2 = np.array([1, 2])
v3 = np.array([4, 2])  # dependent (multiple of v1

#creat plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

plt.suptitle("Visualizing Span and Linear Independence", fontsize=15)

#--- Independent vecttors ---
axes[0].arrow(0, 0, *v1, color='blue', head_width=0.15, length_includes_head=True)

axes[0].arrow(0, 0, *v2, color='green', head_width=0.15, length_includes_head=True)

for a in np.linspace(-1, 1, 10):
    for b in np.linspace(-1, 1, 10):
        point = a*v1 + b*v2
        axes[0].plot(point[0], point[1], 'o', color='gray', alpha=0.2)
axes[0].set_title("Independent Vectors (Span = ℝ²)")
axes[0].set_xlim(-3, 5)
axes[0].set_ylim(-3, 5)
axes[0].grid(True)
axes[0].axhline(0, color='black', linewidth=1)
axes[0].axvline(0, color='black', linewidth=1)
axes[0].legend(['Span Points', 'v1', 'v2'], loc='upper left')

# ---Dependent Vectore---
axes[1].arrow(0, 0, *v1, color='blue', head_width=0.15, length_includes_head=True)
axes[1].arrow(0, 0, *v3, color='red', linestyle='--', head_width=0.15, length_includes_head=True)
axes[1].set_title("Dependent Vectors (Same Direction)")
axes[1].set_xlim(-3, 5)
axes[1].set_ylim(-3, 5)
axes[1].grid(True)
axes[1].axhline(0, color='black', linewidth=1)
axes[1].axvline(0, color='black', linewidth=1)
axes[1].legend(['v1', 'v3'], loc='upper left')

plt.show()
```
